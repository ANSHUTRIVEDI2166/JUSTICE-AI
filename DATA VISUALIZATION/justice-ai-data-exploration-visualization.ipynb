```python
import pandas as pd 

df = pd.read_csv("/kaggle/input/justice-ai-case-dataset/justice_ai_dataset.csv")

df.head()
df.isnull().sum()
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x='case_type', order=df['case_type'].value_counts().index)
plt.xticks(rotation=60)
plt.title("Sorted Case Types")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df,x='severity_level')
plt.xticks(rotation=60)
plt.title("Sorted severity level")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='urgency_level')
sns.countplot(data=df, x='status')
print(df.shape)
print(df['status'].value_counts())


```

```python
sns.countplot(data=df, x='media_attention')
```

```python
sns.countplot(data=df, x='evidence_available')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='victims_count', bins=10)
plt.title('Distribution of Victims Count', fontsize=16)
plt.xlabel('Number of Victims', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
df['date_filed'] = pd.to_datetime(df['date_filed'])
df['date_filed'].dt.year.value_counts().sort_index().plot(kind='bar')
sns.countplot(data=df, x='public_impact_level')
```

```python
plt.figure(figsize=(14, 5))
sns.countplot(data=df, x='case_type', hue='status')
plt.xticks(rotation=45)
plt.title("Case Type vs Status")
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 5))
sns.countplot(data=df, x='case_type', hue='urgency_level')
plt.xticks(rotation=45)
plt.title("Case Type vs Urgency Level")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='status', hue='severity_level')
plt.title("Status vs Severity Level")
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 5))
sns.boxplot(data=df, x='case_type', y='victims_count')
plt.xticks(rotation=45)
plt.title("Victims Count per Case Type")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='status', hue='evidence_available')
plt.title("Evidence Availability Across Case Status")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='case_type', hue='media_attention')
plt.xticks(rotation=45)
plt.title("Case Types with/without Media Attention")
plt.tight_layout()
plt.show()
```
